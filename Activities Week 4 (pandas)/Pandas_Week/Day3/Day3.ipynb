# Instructor Turn Activity 1: Merging 

```python
# Dependencies
import pandas as pd
```

```python
raw_data_info = {
    "customer_id": [112, 403, 999, 543, 123],
    "name": ["John", "Kelly", "Sam", "April", "Bobbo"],
    "email": ["jman@gmail", "kelly@aol.com", "sports@school.edu", "April@yahoo.com", "HeyImBobbo@msn.com"]
}
info_pd = pd.DataFrame(raw_data_info, columns=["customer_id", "name", "email"])
info_pd
```

```python
# Create DataFrames
raw_data_items = {
    "customer_id": [403, 112, 543, 999, 654],
    "item": ["soda", "chips", "TV", "Laptop", "Cooler"],
    "cost": [3.00, 4.50, 600, 900, 150]
}
items_pd = pd.DataFrame(raw_data_items, columns=[
                        "customer_id", "item", "cost"])
items_pd
```

```python
# Merge two dataframes using an inner join, their intersection
merge_table = pd.merge(info_pd, items_pd, on="customer_id")
merge_table
```

```python
# Merge two dataframes using an outer join, both including things that don't intersect
merge_table = pd.merge(info_pd, items_pd, on="customer_id", how="outer")
merge_table
```

```python
# Merge two dataframes using a left join
merge_table = pd.merge(info_pd, items_pd, on="customer_id", how="left")
merge_table
```

```python
# Merge two dataframes using a right join
merge_table = pd.merge(info_pd, items_pd, on="customer_id", how="right")
merge_table
```

## Students Turn Activity 2: Cryptocurrency Merging

### Instructions
  * Read in both of the CSV files and print out their DataFrames.

  * Perform an inner merge that combines both DataFrames on the "Date" column.

  * Rename the columns within the newly merged DataFrame so that the headers are more descriptive.

  * Create a summary table that includes the following information: `Best Bitcoin Open`, `Best Dash Open`, `Best Bitcoin Close`, `Best Dash Close`, `Total Bitcoin Volume`, `Total Dash Volume`.

  * `Total Bitcoin Volume` and `Total Dash Volume` should be calculated to have units of "millions" and be rounded to two decimal places.

```python
# Import Dependencies
import pandas as pd
```

```python
bitcoin_csv = "Resources/bitcoin_cash_price.csv"
dash_csv = "Resources/dash_price.csv"
```

```python
bitcoin_df = pd.read_csv(bitcoin_csv)
dash_df = pd.read_csv(dash_csv)
```

```python
bitcoin_df.head()
```

```python
dash_df.head()
```

```python
# Merge the two DataFrames together based on the Dates they share
merge_table = pd.merge(bitcoin_df, dash_df, on="Date")
merge_table.head()
```

```python
# Rename columns so that they are differentiated
renamed = merge_table.rename(columns={
    "Open_x":"Bitcoin Open", 
    "High_x":"Bitcoin High", 
    "Low_x": "Bitcoin Low",
    "Close_x": "Bitcoin Close",
    "Volume_x": "Bitcoin Volume",
    "Market Cap_x": "Bitcoin Market Cap",
    "Open_y":"Dash Open", 
    "High_y":"Dash High", 
    "Low_y": "Dash Low",
    "Close_y": "Dash Close",
    "Volume_y": "Dash Volume",
    "Market Cap_y": "Dash Market Cap"})
renamed.head()
```

```python
# alternatively you can set your suffixes when the merge occurs
alt_merge = pd.merge(bitcoin_df, dash_df, on="Date", suffixes={" Bitcoin", " Dash"}) #include the space before bitcoin
alt_merge.head()
```

```python
# Collecting best open for Bitcoin and Dash

#WAY EASIER DUH!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!

#bitcoin_open_win = renamed["Bitcoin Open"].max()

#!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!

bitcoin_open_df = renamed.sort_values(["Bitcoin Open"], ascending = False)
bitcoin_open_df.reset_index(inplace = True)
bitcoin_open_win = bitcoin_open_df["Bitcoin Open"][0]
print(bitcoin_open_win)

dash_open_df = renamed.sort_values(["Dash Open"], ascending = False)
dash_open_df.reset_index(inplace = True) 
dash_open_win = dash_open_df["Dash Open"][0]
print(dash_open_win)

# Collecting best close for Bitcoin and Dash
bitcoin_close_df = renamed.sort_values(["Bitcoin Close"], ascending = False)
bitcoin_close_df.reset_index(inplace = True)
bitcoin_close_win = bitcoin_close_df["Bitcoin Close"][0]
print(bitcoin_close_win)

dash_close_df = renamed.sort_values(["Dash Close"], ascending = False)
dash_close_df.reset_index(inplace = True) 
dash_close_win = dash_close_df["Dash Open"][0]
print(dash_close_win)

# Collecting the total volume for Bitcoin and Dash
bitcoin_volume = renamed["Bitcoin Volume"].sum()
print(bitcoin_volume)
dash_volume = renamed["Dash Volume"].sum()
print(dash_volume)

adj_bit_vol = round(renamed["Bitcoin Volume"].sum()/1000000,2)
adj_dash_vol = round(renamed["Dash Volume"].sum()/1000000,2)
print(adj_bit_vol)
print(adj_dash_vol)
```

```python
# Creating a summary DataFrame using above values
#Best Bitcoin Open, Best Dash Open, Best Bitcoin Close, Best Dash Close, Total Bitcoin Volume, Total Dash Volume. 
#Total Bitcoin Volume and Total Dash Volume should be calculated to have units of "millions" and be rounded to two decimal places.


summary = pd.DataFrame({
    "Best Bitcoin Open": [bitcoin_open_win],
    "Best Dash Open": [dash_open_win],
    "Best Bitcoin Close": [bitcoin_close_win],
    "Best Dash Close": [dash_close_win],
    "Total Bitcoin Volume (Millions)": [adj_bit_vol],
    "Total Dash Volume (Millions)": [adj_dash_vol]
})
summary
```

```python
#extra practice - inner, outer, left, and right

import pandas as pd

d1 = {"customer_id": pd.Series([1,2,3,4,5,6]),
     "products": pd.Series(["Oven", "Oven", "Fridge", "TV", "Fridge", "TV"])}
d1_df = pd.DataFrame(d1)

d2 = {"customer_id": pd.Series([2,4,5]),"State": pd.Series(["CA","TX","TX"])}
d2_df = pd.DataFrame(d2)

print("inner join")
print(pd.merge(d1_df, d2_df, on = "customer_id", how = "inner"))

print("outer join")
print(pd.merge(d1_df, d2_df, on = "customer_id", how = "outer"))

print("left join")
print(pd.merge(d1_df, d2_df, on = "customer_id", how = "left"))

print("right join")
print(pd.merge(d1_df, d2_df, on = "customer_id", how = "right"))
```

# Instructor Turn Activity 3 Binning

```python
# Import Dependencies
import pandas as pd
```

```python
raw_data = {
    'Class': ['Oct', 'Oct', 'Jan', 'Jan', 'Oct', 'Jan'], 
    'Name': ["Cyndy", "Logan", "Laci", "Elmer", "Crystle", "Emmie"], 
    'Test Score': [90, 56, 72, 88, 98, 67]}
df = pd.DataFrame(raw_data)
df
```

```python
# Create the bins in which Data will be held
# Bins are 0, 60, 70, 80, 90, 100
bins = [0, 60, 70, 80, 90, 100]

# Create the names for the four bins
group_names = ["F", "D", "C", "B", "A"]
```

```python
df["Test Score Summary"] = pd.cut(df["Test Score"], bins, labels=group_names)
df
```

```python
# Creating a group based off of the bins
df = df.groupby("Test Score Summary")
df.max()
```

```python
# Place the data series into a new column inside of the DataFrame
```

## Students Turn Activity 6: Cleaning Kickstarter

### Instructions

  * The instructions for this activity are contained within the Jupyter Notebook.

```python
import pandas as pd
```

```python
# The path to our CSV file
# The path to our CSV file
file = "Resources/KickstarterData.csv"

# Read our Kickstarter data into pandas
df = pd.read_csv(file)
df.head()
```

```python
# Get a list of all of our columns for easy reference
df.columns
```

```python
# Extract "name", "goal", "pledged", "state", "country", "staff_pick",
# "backers_count", and "spotlight"
new_df = df[["name", "goal", "pledged", "state", "country", "staff_pick", "backers_count", "spotlight"]]
new_df.head()
```

```python
# Remove projects that made no money at all
cleaned_df = new_df.dropna(subset = ['pledged'], how='any')
cleaned_df.head()
#or search for pledged > 0
```

```python
# Collect only those projects that were hosted in the US
new_only_us = cleaned_df.loc[cleaned_df["country"] == "US", :]
new_only_us.head()
```

```python
# Create a new column that finds the average amount pledged to a project
new_only_us["avg_pledged"] = new_only_us["pledged"]/new_only_us["backers_count"]
new_only_us.head()
```

```python
# Format our "average_donation", "goal", and "pledged" columns
# to go to two decimal places, include a dollar sign, and use comma notation

new_only_us['avg_pledged'] = pd.to_numeric(new_only_us['avg_pledged'])

new_only_us["avg_pledged"] = new_only_us["avg_pledged"].map("${:,.2f}".format)

new_only_us["goal"] = new_only_us["goal"].map("${:,.2f}".format)

new_only_us["pledged"] = new_only_us["pledged"].map("${:,.2f}".format)

new_only_us.head()
```

```python
# Calculate the total number of backers for all US projects
new_only_us["backers_count"].sum()
```

```python
# Calculate the average number of backers for all US projects
new_only_us["backers_count"].mean()
```

```python
# Collect only those US campaigns that have been picked as a "Staff Pick"
staff = new_only_us[new_only_us["staff_pick"] == True]
staff.head()
```

```python
# Group by the state of the campaigns and see if staff picks matter (Seems to matter quite a bit)
state_groups = staff.groupby("state")
state_groups
```
